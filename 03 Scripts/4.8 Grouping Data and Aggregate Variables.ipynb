```python
# Create new max_order column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Create flag for customer loyalty 

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check the count for created flag

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check specific columns

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## Task 6

```python
# Calculaten mean price by each user for the product 

ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create spending_flag column 
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Check frequency of new spending_flag column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check updated data frame
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(100)
```

## Task 7

```python
# Calculaten mean day since last order 

ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
# Create frequency flag column 
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] < 10, 'order_frequency_flag'] = 'Frequent Customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Check value counts of new order_frequency column
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check updated data frame
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_since_prior_order', 'order_frequency_flag']].head(25)
```
